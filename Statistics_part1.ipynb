**MEASURE OF CENTRAL TENDENCY**

1. MEAN
2. MEDIAN
3. MODE


```python
ages = [23,24,32,45,12,43,67,45,32,56,32,120]
```

```python
import numpy as np
print(np.mean(ages))
print(np.median(ages)) #There is no impact on median due to any outlier
```

```python
import statistics
print(statistics.mean(ages))
print(statistics.median(ages))
```

```python
print(statistics.mode(ages))
```

```python
import seaborn as sns
sns.boxplot(ages)
```

**5 NUMBER SUMMARY**


```python
import numpy as np
q1,q3=np.percentile(ages,[25,75])
```

```python
print(q1,q3)
```

```python
## To check outlier[Lower fence,higher fence]
IQR=q3-q1
lower_fence=q1-1.5*(IQR)
Higher_fence=q3+1.5*(IQR)
print(lower_fence,Higher_fence)
```

**MEASURE OF** **DISPERSION**

1. VARIANCE
2. STANDARD DEVIATION

```python
print(statistics.variance(ages))
```

```python
np.var(ages,axis=0)
```

```python
def variance(data):
  n=len(ages)
  #mean of the data
  mean=sum(data)/n
  ##variamce
  deviation=[(x - mean)**2 for x in data]
  variance=sum(deviation)/(n-1)
  return variance
```

```python
print(variance(ages))
```

```python
def variance(data,dof=0):
  n=len(ages)
  #mean of the data
  mean=sum(data)/n
  ##variamce
  deviation=[(x - mean)**2 for x in data]
  variance=sum(deviation)/(n-dof)
  return variance
```

```python
print(variance(ages))
```

```python
statistics.pvariance(ages)
```

```python
import math
math.sqrt(statistics.pvariance(ages))
```

**Check whether Distribution is normal Distribution**

```python
##if you want to check whether feature is guassian or normal distribution
###   Q-Q plot

import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  sns.histplot(sample)
  plt.subplot(1,2,2)
  stat.probplot(sample,dist='norm',plot=pylab)
  plt.show()
```

```python
## create a normal distributed data
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
```

```python
plot_data(df['sepal_width'])
```

```python
mu,sigma = 3., 1. #mean and sd
sample = np.random.lognormal(mu,sigma,100)
```

```python
def plot_data_log(sample):##lognormal distribution
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  sns.histplot(sample)
  plt.subplot(1,2,2)
  stat.probplot(np.log(sample),dist='norm',plot=pylab)
  plt.show()
```

```python
plot_data_log(sample)
```
